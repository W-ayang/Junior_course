```matlab
I = imread('\lena.jpg');

[M,N,p] = size(I);

g = zeros(M,N);
a = 40;b = 40;
for i = 1:M
    for j=1:N
        if((i - a > 0) & (i - a <M) & (j - b >0) &(j - b < N))
            g(i,j) = I(i - a,j - b);
        else
            g(i ,j) = 0;
        end
    end
end
figure(1);imshow(uint8(g));
```

```matlab
I = imread('\lena.jpg');

subplot(1,2,1);
imshow(I);
title('');

I = im2double(I);
J = fft2(I);
K = fftshift(J);
K = abs(K);

subplot(1,2,2);
imshow(log(K + 1),[]);
title('');
```

```matlab
I1 = zeros(256,256);
I1(28:228,120:156) = 1;
subplot(1,4,1)
imshow(I1);

I1 = im2double(I1);
J1 = fft2(I1);
K1 = fftshift(J1);
K1 = abs(K1);
subplot(1,4,2);
imshow(K1);

I2 = imrotate(I1,315,"bilinear","crop");
subplot(1,4,3);
imshow(I2);

I2 = im2double(I2);
J2 = fft2(I2);
K2 = fftshift(J2);
K2 = abs(K2);

subplot(1,4,4);
imshow(K2);


```

```matlab
subplot(1,2,1);
I1 = imshow("lena.jpg");
% I1 = im2double(I1);

[M,N] = size(I1);

I2 = zeros(M,N);

for i=1:M
    for j=1:N
        I2(i,j) = I1(i,j) * power(-1,i+j);
    end
end

J2 = im2double(I2);
J2 = fft2(I2);
K2 = fftshift(J2);
K2 = abs(K2);

subplot(1,2,2);
imshow(K2,[]);
```
