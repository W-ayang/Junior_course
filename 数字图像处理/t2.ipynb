```matlab
I = imread('\lena.jpg');
J = imnoise(I,'gaussian',0,0.02);

Q = J - I;
Q = 255 - Q;
subplot(1,3,1);imshow(I);
subplot(1,3,2);imshow(J);
subplot(1,3,3);imshow(Q);
```

```matlab
I = imread('\lena.jpg');

[M,N,p] = size(I);

g = zeros(M,N);
a = 40;b = 40;
for i = 1:M
    for j=1:N
        if((i - a > 0) & (i - a <M) & (j - b >0) &(j - b < N))
            g(i,j) = I(i - a,j - b);
        else
            g(i ,j) = 0;
        end
    end
end
figure(1);imshow(uint8(g));
```

```matlab
I = imread('\lena.jpg');
subplot(1,2,1);imshow(I);title('');
[M,N,p] = size(I);

g = zeros(M,N);

for i=1:M
    for j=1:N
        g(i,j) = I(i,N+1-j);
    end
end
subplot(1,2,2);imshow(uint8(g));title('')

```
