```matlab
x = imread('数字图像处理\lena.jpg');
x = im2double(x);
figure(1);
subplot(4,2,1);
imshow(x);
title('原始图像')

a = 2;b = -50;
y = a.*x + b/255;
figure(1);
subplot(4,2,2);
imshow(y);
title('增加对比度')

a = 0.5;b = -50;
w = a.* x + b/255;
figure(1);
subplot(4,2,3);
imshow(w)
title('减小对比度');

a = 1;b = 50;
q = a .* x + b/255;
subplot(4,2,4);
imshow(q)
title('线性增加亮度');

a = -0.5;b = 255;
r = a.* x + b/255;
subplot(4,2,5);
imshow(r);
title('反色');

% 非线性变换
s = 1.1 * log2(1 + x);
subplot(4,2,6);
imshow(s)

c = 1.1;
gamma = 0.25;
s2 = c*x.^gamma;
figure(1);
subplot(4,2,7);
imshow(s2);
title('幂次变换');

```
