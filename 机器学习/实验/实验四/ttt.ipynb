```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 

# 读取鸢尾花数据集
data_df = pd.read_csv('D:\code\大三课程\机器学习\实验\实验四\iris.csv')

# 提取特征和类别
X = data_df.iloc[:, :-1].values  # 特征
y = data_df.iloc[:, -1].values   # 类别

# 创建一个散点图，根据不同类别使用不同颜色标记数据点
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 'setosa', 0], X[y == 'setosa', 1], label='Setosa', c='red', marker='o')
plt.scatter(X[y == 'versicolor', 0], X[y == 'versicolor', 1], label='Versicolor', c='blue', marker='s')
plt.scatter(X[y == 'virginica', 0], X[y == 'virginica', 1], label='Virginica', c='green', marker='^')

# 添加图像右上角标签
plt.xlabel('花萼长度')
plt.ylabel('花萼宽度')
plt.title('鸢尾花数据集的散点图')
plt.legend(loc='upper right')

# 显示图形
plt.show()
```
