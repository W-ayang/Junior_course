```matlab
% %% 1.二次函数
% % 示例1
% [X1,Y1] = meshgrid(-3:0.5:3); 
% Z1 = X1.^2 + Y1.^2;
% mesh(X1,Y1,Z1);

% 示例2
[X2,Y2] = meshgrid(-2:0.2:2);
Z2 = 0.5*X2.^2 + 2*Y2.^2; 
mesh(X2,Y2,Z2);
```

```matlab
%% 2.正弦函数
% 示例1
[X1,Y1] = meshgrid(-2:0.1:2);
Z1 = sin(X1) + cos(Y1);
mesh(X1,Y1,Z1);

% % 示例2
% [X2,Y2] = meshgrid(-3:0.2:3);
% Z2 = 3*sin(X2) + 2*cos(Y2);
% mesh(X2,Y2,Z2);
```

```matlab
%% 3.指数函数
% 示例1 
[X1,Y1] = meshgrid(-3:0.5:3);
Z1 = exp(-X1.^2 - Y1.^2);
mesh(X1,Y1,Z1);

% 示例2
[X2,Y2] = meshgrid(-3:0.2:3);
Z2 = 2*exp(-0.5*(X2 + Y2));
mesh(X2,Y2,Z2);
```

```matlab
%% 4.对数函数
% 示例1
[X1,Y1] = meshgrid(-2:0.2:2); 
Z1 = log(abs(X1))+log(abs(Y1));
mesh(X1,Y1,Z1);

% 示例2
[X2,Y2] = meshgrid(-3:0.5:3);
Z2 = log(1+X2.^2+Y2.^2);
mesh(X2,Y2,Z2);
```

```matlab
%% 5.多项式函数
% 示例1
[X1,Y1] = meshgrid(-2:0.1:2);
Z1 = X1.^2 + 2*X1.*Y1 + Y1.^3; 
mesh(X1,Y1,Z1);

% 示例2  
[X2,Y2] = meshgrid(-3:0.2:3);
Z2 = 2*X2.^3 - 3*X2 + Y2.^2;
mesh(X2,Y2,Z2);
```

```matlab
%% 6.分段函数
% 示例1
[X1,Y1] = meshgrid(-3:0.1:3);
Z1 = double(X1 > 0); 
mesh(X1,Y1,Z1);

% 示例2
[X2,Y2] = meshgrid(-3:0.2:3);
Z2 = double(Y2 > X2);
mesh(X2,Y2,Z2);
```

```matlab
%% 7.三角函数
% 示例1
[X1,Y1] = meshgrid(-2:0.1:2);  
Z1 = tan(X1) + cot(Y1);
mesh(X1,Y1,Z1);

% 示例2
[X2,Y2] = meshgrid(-3:0.2:3);
Z2 = sec(X2).*csc(Y2);
mesh(X2,Y2,Z2);
```

```matlab
%% 8.导函数
% 示例1
[X1,Y1] = meshgrid(-2:0.1:2);
Z1 = gradient(X1.^2+Y1.^2);  
mesh(X1,Y1,Z1);

% 示例2
[X2,Y2] = meshgrid(-3:0.2:3); 
Z2 = gradient(sin(X2.*Y2));
mesh(X2,Y2,Z2);
```
